## np.round(i, 반올리함려는 자리수)




```python
import math

a = 6.123456789012345

a1 = math.ceil(a)  # 소수점 1의 자리에서 올림

a4 = math.ceil(a * 10 ** 4) / 10 ** 4    #소수점 4의 자리에서 올림

a5 = np.round(a, 3)    

print (a1)
print (a4)
print (a5)
```
